### Proyecto de simulación de inteligencia artificial  utiliza POO, tipo Menu
### 18. Diseñador: Portafolio digital, software de diseño, estilo artístico, clientes activos, tarifa por proyecto y tipo de hardware.

```python
# --- DEFINICIÓN DE LA CLASE [6-8] ---
class Diseñador:
    """Clase que representa a un profesional del diseño [3, 8]."""

    def __init__(self, nombre, portafolio, software, estilo, clientes, tarifa, hardware):
        """Constructor para inicializar los atributos del diseñador [9-11]."""
        self.nombre = nombre
        self.portafolio = portafolio
        self.software = software
        self.estilo = estilo
        self.clientes = int(clientes)  # Se convierte a entero
        self.tarifa = float(tarifa)    # Se convierte a flotante
        self.hardware = hardware

    def analizar_potencial_ia(self):
        """Método que simula una lógica de IA para evaluar al diseñador [12, 13]."""
        if self.clientes > 8 and self.tarifa > 500:
            return "ELITE: Alta rentabilidad detectada."
        elif self.clientes > 0:
            return "ACTIVO: Carga de trabajo estable."
        else:
            return "PASIVO: Potencial de crecimiento disponible."

# --- INICIALIZACIÓN DEL SISTEMA ---
lista_diseñadores = [] # Lista para almacenar objetos de la clase Diseñador [14]

# --- LÓGICA INTERACTIVA (MENÚ) ---
print("====================================================")
print("     NÚCLEO DE INTELIGENCIA ARTIFICIAL POO v2.0     ")
print("====================================================")

while True:
    # Lógica de Percepción de la IA sobre la lista general
    cantidad = len(lista_diseñadores)
    if cantidad == 0:
        estado_ia = "IA: Esperando inicialización de perfiles."
    else:
        estado_ia = f"IA: Monitoreando {cantidad} instancias en el sistema."

    print(f"\n[{estado_ia}]")
    print("1. Registrar Instancia (Nuevo Diseñador)")
    print("2. Visualizar Tabla de Datos (Objetos)")
    print("3. Analizar Perfiles con IA")
    print("4. Apagar Núcleo")

    opcion = input("\nInstrucción de usuario: ")

    if opcion == "1":
        print("\n>>> CAPTURA DE ATRIBUTOS DEL OBJETO:")
        # Entrada de datos interactiva [6, 15]
        nom = input("Nombre: ")
        port = input("Portafolio Digital (URL): ")
        soft = input("Software de diseño: ")
        est = input("Estilo artístico: ")
        cli = input("Clientes activos: ")
        tar = input("Tarifa por proyecto: ")
        hard = input("Tipo de hardware: ")

        # INSTANCIACIÓN: Se crea el objeto y se añade a la lista [7, 9, 16]
        nuevo_objeto = Diseñador(nom, port, soft, est, cli, tar, hard)
        lista_diseñadores.append(nuevo_objeto)
        print(f"\n>>> IA: Objeto '{nom}' instanciado y almacenado en memoria.")

    elif opcion == "2":
        if len(lista_diseñadores) > 0:
            # Encabezado de tabla alineado [17, 18]
            print("\n" + "-"*135)
            print(f"{'NOMBRE':<15} | {'SOFTWARE':<15} | {'ESTILO':<15} | {'CLIENTES':<10} | {'TARIFA':<10} | {'HARDWARE':<15} | {'PORTAFOLIO':<25}")
            print("-" * 135)

            # CICLO FOR para recorrer los objetos de la lista [19, 20]
            for d in lista_diseñadores:
                # Acceso a atributos mediante notación de punto (.) [13, 21, 22]
                print(f"{d.nombre:<15} | {d.software:<15} | {d.estilo:<15} | {d.clientes:<10} | ${d.tarifa:<9.2f} | {d.hardware:<15} | {d.portafolio:<25}")
            print("-" * 135)
        else:
            print("\n>>> IA: No existen instancias para generar la tabla.")

    elif opcion == "3":
        if len(lista_diseñadores) > 0:
            print("\n>>> IA REALIZANDO ANÁLISIS DE COMPORTAMIENTO:")
            for d in lista_diseñadores:
                # Llamada a función interna del objeto [23, 24]
                resultado = d.analizar_potencial_ia()
                print(f"- Diseñador {d.nombre}: {resultado}")
        else:
            print("\n>>> IA: Memoria vacía. Nada que analizar.")

    elif opcion == "4":
        print("\n>>> IA: Desconectando procesos POO... Sistema fuera de línea.")
        break # Rompe el ciclo para salir [25]

    else:
        print("\n>>> IA ERROR: Comando inválido.")
```
