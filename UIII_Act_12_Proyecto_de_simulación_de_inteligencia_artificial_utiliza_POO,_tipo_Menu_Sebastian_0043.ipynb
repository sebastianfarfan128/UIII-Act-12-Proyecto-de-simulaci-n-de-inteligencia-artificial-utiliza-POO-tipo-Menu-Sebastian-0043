### Proyecto de simulación de inteligencia artificial  utiliza POO, tipo Menu
### 18. Diseñador: Portafolio digital, software de diseño, estilo artístico, clientes activos, tarifa por proyecto y tipo de hardware.

```python
# --- DEFINICIÓN DE LA CLASE ---
class Diseñador:
    """Representa a un profesional del diseño en el sistema de la IA [1, 5]."""
    def __init__(self, nombre, portafolio, software, estilo, clientes, tarifa, hardware):
        # Atributos de la instancia [2, 6]
        self.nombre = nombre
        self.portafolio = portafolio
        self.software = software
        self.estilo = estilo
        self.clientes = int(clientes)
        self.tarifa = float(tarifa)
        self.hardware = hardware

# --- BASE DE DATOS Y FUNCIONES DEL SISTEMA ---
los_diseñadores = [] # Lista para almacenar objetos de la clase Diseñador [7]

def agregar_diseñador():
    """Crea una nueva instancia de Diseñador y la añade a la lista [8]."""
    print("\n>>> IA: Iniciando captura de datos del nuevo perfil.")
    nom = input("Nombre: ")
    port = input("URL Portafolio: ")
    soft = input("Software principal: ")
    est = input("Estilo artístico: ")
    cli = input("Clientes activos: ")
    tar = input("Tarifa por proyecto (USD): ")
    hard = input("Tipo de hardware: ")

    # INSTANCIACIÓN: Creación del objeto [9]
    nuevo = Diseñador(nom, port, soft, est, cli, tar, hard)
    los_diseñadores.append(nuevo)
    print(f"\n>>> IA: Objeto '{nom}' integrado con éxito.")

def listar_diseñadores():
    """Muestra los objetos almacenados en formato de tabla alineada [10]."""
    if len(los_diseñadores) == 0:
        print("\n>>> IA: No existen registros para listar.")
    else:
        print("\n" + "="*145)
        print(f"{'NOMBRE':<15} | {'SOFTWARE':<15} | {'ESTILO':<15} | {'CLIENTES':<10} | {'TARIFA':<10} | {'HARDWARE':<15} | {'PORTAFOLIO':<30}")
        print("-" * 145)
        # Recorrido de objetos mediante ciclo for [8, 11]
        for d in los_diseñadores:
            print(f"{d.nombre:<15} | {d.software:<15} | {d.estilo:<15} | {d.clientes:<10} | ${d.tarifa:<9.2f} | {d.hardware:<15} | {d.portafolio:<30}")
        print("="*145)

def actualizar_diseñador():
    """Busca un objeto por nombre y modifica sus atributos [12]."""
    nombre_buscado = input("\n>>> IA: Ingrese el nombre del diseñador a actualizar: ")
    encontrado = False
    for d in los_diseñadores:
        if d.nombre.lower() == nombre_buscado.lower():
            print(f">>> IA: Perfil localizado. Ingrese nuevos valores para {d.nombre}:")
            d.clientes = int(input("Nuevos clientes activos: "))
            d.tarifa = float(input("Nueva tarifa por proyecto: "))
            d.hardware = input("Nuevo hardware: ")
            print(">>> IA: Atributos del objeto actualizados.")
            encontrado = True
    if not encontrado:
        print(">>> IA: No se encontró coincidencia en la memoria.")

def borrar_diseñador():
    """Elimina una instancia de la lista principal [13, 14]."""
    nombre_buscado = input("\n>>> IA: Nombre del perfil a eliminar de la base de datos: ")
    for i in range(len(los_diseñadores)):
        if los_diseñadores[i].nombre.lower() == nombre_buscado.lower():
            eliminado = los_diseñadores.pop(i)
            print(f">>> IA: Objeto '{eliminado.nombre}' destruido correctamente.")
            return
    print(">>> IA: Error. Perfil no localizado.")

# --- BUCLE PRINCIPAL (MENÚ INTERACTIVO) ---
while True:
    # Lógica de simulación de IA (Reporte de estado) [7]
    cantidad = len(los_diseñadores)
    if cantidad == 0:
        estado_ia = "PASIVO: Esperando datos."
    elif cantidad < 3:
        estado_ia = "ÓPTIMO: Escaneo de talento activo."
    else:
        estado_ia = "CRÍTICO: Alta carga de registros detectada."

    print(f"\n[ESTADO IA]: {estado_ia} (Total: {cantidad})")
    print("1. Agregar Diseñador")
    print("2. Listar Diseñadores (Tabla)")
    print("3. Actualizar Diseñador")
    print("4. Borrar Diseñador")
    print("5. Apagar Núcleo IA")

    opcion = input("\nComando: ")

    if opcion == "1":
        agregar_diseñador()
    elif opcion == "2":
        listar_diseñadores()
    elif opcion == "3":
        actualizar_diseñador()
    elif opcion == "4":
        borrar_diseñador()
    elif opcion == "5":
        print("\n>>> IA: Desconectando protocolos POO... Adiós.")
        break
    else:
        print("\n>>> IA ERROR: Comando no identificado.")
```
